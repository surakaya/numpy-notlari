# **NumPy ile Keşifçi Veri Analizi (EDA)**

NumPy, Python'da bilimsel hesaplamalar yapmak için kullanılan güçlü bir kütüphanedir. Özellikle çok boyutlu diziler ve matrislerle çalışmak için tasarlanmıştır ve matematiksel fonksiyonlar sağlar. 

```python
import numpy as np #kütüphanemizi projemize import ettik.
```

**NumPy'da Diziler**

```python
listem = [1,2,3,4,5]
type(listem) #dizimiz liste tipindedir.
```

```python
listem = np.array(listem) #numpy dizisine dönüştürdük.
print(listem.ndim)
```

```python
#çok boyutlu numpy dizisi oluşturma:
ikiBoyutluDizi = np.array([ [1,2,3], [4,5,6] ]) 
ikiBoyutluDizi
```

**Diziler üzerinde temel işlemler**

```python
yeniDizi1 = np.arange(0,10)
#TOPLAMA :
print( f" Dizideki elemanların toplamı: {yeniDizi1.sum()}"  )
```

```python
yeniDizi2 = np.arange(0,10)
#ORTALAMA :
print( f" Dizideki elemanların ortalaması: {yeniDizi2.mean()}"  )
```

```python
yeniDizi3 = np.arange(0,10)
#MAKSİMUM :
print( f" Dizideki elemanların EN BÜYÜĞÜ: {yeniDizi3.max()}"  )
```

```python
yeniDizi4 = np.arange(0,10)
#STANDART SAPMA :
print( f" Dizideki elemanların standart sapması: {yeniDizi4.std()}"  )
```

```python
#diziyi tersine çevirmek:
listem[::-1]
```

**Çeşitli dizileri oluşturma yöntemleri**

```python
#arange(<başlangıç>, <bitiş>, <kaç farkla artacak>) 
np.arange(0,10,2)
```

```python
#np.zeros(x) -> x adet 0'dan oluşan dizi :
np.zeros(5, dtype = int)
```

```python
#np.ones(y) -> y adet 1'den oluşan dizi :
np.ones(7, dtype = int)
```

```python
np.full ((3,3),5) #3x3'lük, 5 sayısından oluşan matris
```

```python
# np.linspace(a,b,c) -> a ile b arasında, eşit aralıklı c adet sayı içeren dizi :
np.linspace(0,20,5)
```

```python
# np.eye(z) -> z*z'lik köşegen matrisi oluşturmak :
np.eye(7)
```

```python
# np.random.randn(a) -> a adet rastgele değerden oluşan dizi :
np.random.randn(5)
```

```python
# np.random.randint(b,c,d) -> b ile c arasında rastgele d adet sayı döndürmek :
np.random.randint(4,9,10)
```

```python
# 0-40 arası değerlerden oluşan diziyi 4 sütun-10 satıra bölmek :
dizim = np.arange(0,40)
dizim.reshape(4,10)
```

**NumPy'da Indexleme**

```python
benimDizim = np.arange(0,15)
benimDizim[3] # yoluyla istenilen elemana erişilir.
```

```python
#benimDizim[a:b] -> a'dan b'ye kadar olan elemanları basar. (a dahil,b değil)
benimDizim[3:7]
```

**Slicing İşlemi**

```python
baskaDizi = np.arange(0,24)
slicingDizisi = baskaDizi[4:9]
slicingDizisi[:] = 700
slicingDizisi
#2. satırda seçtiğimiz aralıkta bulunan elemanları 3. satırda verdiğimiz değerle değiştirebiliriz.
```
